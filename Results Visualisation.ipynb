```python
import numpy as np
import matplotlib.pyplot as plt

data = [[0.07064, 0.20335, 0.01164], [0.15505, 0.38125, 0.14436]]

X = np.arange(3)
# fig = plt.figure()
# ax = fig.add_axes([0,0,1,1])

plt.bar(X, data[0], width = 0.25, label='Without Bounding Boxes')
plt.bar(X + 0.25, data[1], width = 0.25, label='With Bounding Boxes')
# ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
# ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
# ax.legend(labels=['Without Bounding Boxes', 'With Bounding Boxes'])
# ax.set_xticks(X, ['${AP}^{IoU=.50:.05:.95}$', '${AP}^{IoU=.50}$', '${AP}^{IoU=.75}$'])
plt.xticks(X + 0.25 / 2, ('${AP}^{IoU=.50:.05:.95}$', '${AP}^{IoU=.50}$', '${AP}^{IoU=.75}$'))
plt.legend(loc='best')
plt.xticks(fontsize=12)
plt.title("${AP}^{IoU=.50:.05:.95}$, ${AP}^{IoU=.50}$ and ${AP}^{IoU=.75}$ for Bounding Box Detection", y=1.02)
plt.savefig("AP_1")
```
